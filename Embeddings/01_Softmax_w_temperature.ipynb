```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Softmax with temperature

This notebook implements softmax function as proposed in a paper "Distilling the Knowledge in a Neural Network" by Hinton et al., 2005

```python
logits = np.random.rand(50) * 5
```

```python
def softmax_with_temperature(logits, temp = 1):
    return np.exp(logits / temp) / np.sum(np.exp(logits / temp))
```

"Using higher value of $T$ produces a softer probability distribution over classes" - Hinton et al., 2005

```python
plt.figure(figsize = (7, 7))

for t in range(1, 6, 2):
    distr = softmax_with_temperature(logits, t)
    plt.hist(distr, 
             alpha = .3, 
             label = f'$T$ = {t}',
             bins  = 25)
plt.legend()
plt.title('Temperature vs distribution')
plt.xlabel('$P$')
plt.ylabel('Frequency')
plt.show()
```
